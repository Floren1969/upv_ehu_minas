```python
import numpy as np
```

```python
X=np.linspace(-4*np.pi,4*np.pi,200,endpoint=True)
C,S=np.cos(X),np.sin(X)
```

```python
plt.plot(X,C,'+',color='red',linestyle='-',label="$\cos(x)$")
plt.plot(X,S,':',color='blue',linewidth=5,label="$\sin(x)$")
plt.xlabel('medias',size=20)
plt.xticks([-2*np.pi,0,2*np.pi],[r'$-2 \pi$',r'$2 \pi$'],size=12)
plt.xlim(-10,10)
plt.ylabel('alturas')
plt.yticks([-1,0,1],[r'$-1$',r'$0$',r'$1$'],size=12)
plt.ylim(-1,1)
plt.title(r'GRAFICO $x^2$',size=30)
plt.grid(True)
plt.legend(loc=2) # 0 1 2 3 4 5 6.....10
# lo siguiente es para modificar los cuatro lados de la figura SPINES
ax=plt.gca()
ax.spines['right'].set_color('red')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('right')
```

```python
plt.figure(figsize=(20,6),dpi=120) # es curioso que el 20 es largo y el 6 es ancho ser todo en inch
x = np.arange(0.0, 2, 0.01)
y1 = np.sin(2*np.pi*x)
y2 = 1.2*np.sin(4*np.pi*x)
plt.subplot(1,3,1)
plt.fill_between(x, 0, y1,color='r')
ax1=plt.gca()
ax1.set_ylabel('between y1 and 0')
ax1.set_xlabel('x')
plt.subplot(1,3,2)
plt.plot(x,np.zeros(len(x)),':',color='b')
plt.fill_between(x, y1, y2,color='orange')
ax2=plt.gca()
ax2.set_ylabel('between y1 and y2')
ax1.set_xlabel('x')
plt.subplot(1,3,3)
plt.fill_between(x, 1, y2,color='black')
ax3=plt.gca()
ax3.set_ylabel('between 1 and y2')
ax3.set_xlabel('x')
```

```python
plt.fill_between(x, y1, y2, where=y2 >= y1, facecolor='green', interpolate=True)
plt.fill_between(x, y1, y2, where=y2 <= y1, facecolor='red', interpolate=True)
ax=plt.gca()
ax.set_title('fill between y1 e y2 con diferentes colores usando :where')
```
